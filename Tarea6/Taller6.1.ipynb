**Parte 1: Determinaci贸n del n煤mero 贸ptimo de clusters**

```python
# Instalar librer铆as necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from scipy.spatial.distance import cdist
import seaborn as sns

# Configuraci贸n de estilo profesional
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = '#f8f9fa'
plt.rcParams['font.size'] = 11
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['figure.titlesize'] = 16

# Cargar el dataset de Iris
iris = load_iris()
X = iris.data

# Estandarizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 1. M茅todo del Codo de Jamb煤
def plot_elbow_method(X):
    distortions = []
    K = range(1, 10)
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        distortions.append(sum(np.min(cdist(X, kmeans.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # L铆nea principal con gradiente de color
    line = ax.plot(K, distortions, color='#3498db', linewidth=3, 
                   marker='o', markersize=10, markerfacecolor='#e74c3c',
                   markeredgecolor='white', markeredgewidth=2,
                   label='Distorsi贸n')
    
    # rea bajo la curva
    ax.fill_between(K, distortions, alpha=0.2, color='#3498db')
    
    # Resaltar el codo 贸ptimo (k=3)
    optimal_k = 3
    ax.axvline(x=optimal_k, color='#2ecc71', linestyle='--', linewidth=2.5, 
               alpha=0.7, label=f'k 贸ptimo sugerido = {optimal_k}')
    ax.scatter([optimal_k], [distortions[optimal_k-1]], s=400, 
               color='#2ecc71', edgecolors='white', linewidth=3, 
               zorder=5, marker='*', label='Punto del codo')
    
    ax.set_xlabel('N煤mero de Clusters (k)', fontweight='bold', fontsize=13)
    ax.set_ylabel('Distorsi贸n Promedio', fontweight='bold', fontsize=13)
    ax.set_title('M茅todo del Codo de Jamb煤 (Elbow Method)\nDeterminar el n煤mero 贸ptimo de clusters', 
                 fontweight='bold', fontsize=15, pad=20)
    
    ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.8)
    ax.legend(loc='upper right', framealpha=0.95, edgecolor='black', fontsize=11)
    ax.set_xticks(K)
    
    # Anotaciones en puntos clave
    for i, (k, dist) in enumerate(zip(K, distortions)):
        if k in [1, 3, 5, 9]:
            ax.annotate(f'{dist:.3f}', xy=(k, dist), xytext=(0, 10),
                       textcoords='offset points', ha='center',
                       fontsize=9, fontweight='bold',
                       bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.7))
    
    plt.tight_layout()
    plt.show()
    
    return distortions

# 2. Coeficiente de Silueta
def plot_silhouette_scores(X):
    silhouette_scores = []
    K = range(2, 10)
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        silhouette_scores.append(silhouette_score(X, kmeans.labels_))
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Crear barras con gradiente de color
    colors = plt.cm.RdYlGn(np.linspace(0.3, 0.9, len(K)))
    bars = ax.bar(K, silhouette_scores, color=colors, edgecolor='black', 
                  linewidth=1.5, alpha=0.8)
    
    # Resaltar el mejor valor
    best_k = K[np.argmax(silhouette_scores)]
    best_score = max(silhouette_scores)
    bars[np.argmax(silhouette_scores)].set_color('#2ecc71')
    bars[np.argmax(silhouette_scores)].set_edgecolor('darkgreen')
    bars[np.argmax(silhouette_scores)].set_linewidth(3)
    
    # L铆nea de tendencia
    ax.plot(K, silhouette_scores, color='#e74c3c', linewidth=2.5, 
            marker='D', markersize=8, markerfacecolor='white',
            markeredgecolor='#e74c3c', markeredgewidth=2,
            linestyle='--', alpha=0.7, label='Tendencia')
    
    ax.axhline(y=0.5, color='gray', linestyle=':', linewidth=2, 
               alpha=0.5, label='Umbral aceptable (0.5)')
    
    ax.set_xlabel('N煤mero de Clusters (k)', fontweight='bold', fontsize=13)
    ax.set_ylabel('Coeficiente de Silueta', fontweight='bold', fontsize=13)
    ax.set_title('Coeficiente de Silueta\nMayor valor indica mejor separaci贸n de clusters', 
                 fontweight='bold', fontsize=15, pad=20)
    
    ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.8, axis='y')
    ax.legend(loc='lower right', framealpha=0.95, edgecolor='black', fontsize=11)
    ax.set_xticks(K)
    ax.set_ylim([0, 1])
    
    # Anotaciones en las barras
    for i, (k, score) in enumerate(zip(K, silhouette_scores)):
        ax.text(k, score + 0.02, f'{score:.3f}', ha='center', 
                fontsize=9, fontweight='bold')
    
    # Destacar el mejor
    ax.annotate(f'Mejor k={best_k}\nScore={best_score:.3f}', 
                xy=(best_k, best_score), xytext=(best_k+1.5, best_score-0.15),
                fontsize=11, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.8', facecolor='#2ecc71', alpha=0.8),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.3',
                              color='darkgreen', lw=2))
    
    plt.tight_layout()
    plt.show()
    
    return silhouette_scores

# 3. ndice de Davies-Bouldin
def plot_davies_bouldin_scores(X):
    db_scores = []
    K = range(2, 10)
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        db_scores.append(davies_bouldin_score(X, kmeans.labels_))
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # L铆nea principal con 谩rea
    line = ax.plot(K, db_scores, color='#9b59b6', linewidth=3, 
                   marker='s', markersize=10, markerfacecolor='#e67e22',
                   markeredgecolor='white', markeredgewidth=2,
                   label='ndice DB')
    
    ax.fill_between(K, db_scores, alpha=0.2, color='#9b59b6')
    
    # Resaltar el mejor valor (m铆nimo)
    best_k = K[np.argmin(db_scores)]
    best_score = min(db_scores)
    ax.scatter([best_k], [best_score], s=500, 
               color='#2ecc71', edgecolors='white', linewidth=3, 
               zorder=5, marker='*', label=f'ptimo k={best_k}')
    
    ax.axvline(x=best_k, color='#2ecc71', linestyle='--', linewidth=2.5, 
               alpha=0.7)
    
    ax.set_xlabel('N煤mero de Clusters (k)', fontweight='bold', fontsize=13)
    ax.set_ylabel('ndice de Davies-Bouldin', fontweight='bold', fontsize=13)
    ax.set_title('ndice de Davies-Bouldin\nMenor valor indica mejor separaci贸n de clusters', 
                 fontweight='bold', fontsize=15, pad=20)
    
    ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.8)
    ax.legend(loc='upper right', framealpha=0.95, edgecolor='black', fontsize=11)
    ax.set_xticks(K)
    
    # Anotaciones
    for i, (k, score) in enumerate(zip(K, db_scores)):
        if k == best_k or k in [2, 9]:
            ax.annotate(f'{score:.3f}', xy=(k, score), xytext=(0, -15),
                       textcoords='offset points', ha='center',
                       fontsize=9, fontweight='bold',
                       bbox=dict(boxstyle='round,pad=0.5', facecolor='lightyellow', alpha=0.8))
    
    plt.tight_layout()
    plt.show()
    
    return db_scores

# 4. ndice de Dunn
def calculate_dunn_index(X, labels):
    distances = cdist(X, X, metric='euclidean')
    inter_cluster_distances = []
    intra_cluster_distances = []

    for i in range(np.max(labels) + 1):
        points_in_cluster = np.where(labels == i)[0]
        if len(points_in_cluster) > 1:
            intra_cluster_distances.append(np.mean(distances[np.ix_(points_in_cluster, points_in_cluster)]))

    for i in range(np.max(labels) + 1):
        for j in range(i + 1, np.max(labels) + 1):
            points_in_cluster_i = np.where(labels == i)[0]
            points_in_cluster_j = np.where(labels == j)[0]
            inter_cluster_distances.append(np.mean(distances[np.ix_(points_in_cluster_i, points_in_cluster_j)]))

    return np.min(inter_cluster_distances) / np.max(intra_cluster_distances)

def plot_dunn_index(X):
    dunn_scores = []
    K = range(2, 10)
    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        dunn_scores.append(calculate_dunn_index(X, kmeans.labels_))
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Crear barras con gradiente
    colors = plt.cm.plasma(np.linspace(0.2, 0.8, len(K)))
    bars = ax.bar(K, dunn_scores, color=colors, edgecolor='black', 
                  linewidth=1.5, alpha=0.8)
    
    # Resaltar el mejor valor (m谩ximo)
    best_k = K[np.argmax(dunn_scores)]
    best_score = max(dunn_scores)
    bars[np.argmax(dunn_scores)].set_color('#2ecc71')
    bars[np.argmax(dunn_scores)].set_edgecolor('darkgreen')
    bars[np.argmax(dunn_scores)].set_linewidth(3)
    
    # L铆nea de tendencia
    ax.plot(K, dunn_scores, color='#c0392b', linewidth=2.5, 
            marker='o', markersize=8, markerfacecolor='white',
            markeredgecolor='#c0392b', markeredgewidth=2,
            linestyle='--', alpha=0.7, label='Tendencia')
    
    ax.set_xlabel('N煤mero de Clusters (k)', fontweight='bold', fontsize=13)
    ax.set_ylabel('ndice de Dunn', fontweight='bold', fontsize=13)
    ax.set_title('ndice de Dunn\nMayor valor indica clusters mejor separados y compactos', 
                 fontweight='bold', fontsize=15, pad=20)
    
    ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.8, axis='y')
    ax.legend(loc='upper right', framealpha=0.95, edgecolor='black', fontsize=11)
    ax.set_xticks(K)
    
    # Anotaciones en las barras
    for i, (k, score) in enumerate(zip(K, dunn_scores)):
        ax.text(k, score + max(dunn_scores)*0.02, f'{score:.3f}', 
                ha='center', fontsize=9, fontweight='bold')
    
    # Destacar el mejor
    ax.annotate(f'Mejor k={best_k}\nScore={best_score:.3f}', 
                xy=(best_k, best_score), xytext=(best_k-1.5, best_score*0.75),
                fontsize=11, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.8', facecolor='#2ecc71', alpha=0.8),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.3',
                              color='darkgreen', lw=2))
    
    plt.tight_layout()
    plt.show()
    
    return dunn_scores

# Ejecutar las funciones y mostrar gr谩ficos
print("\n" + "="*80)
print(" ANLISIS DE NMERO PTIMO DE CLUSTERS - DATASET IRIS")
print("="*80 + "\n")

print(" Generando visualizaciones de los 4 m茅todos de evaluaci贸n...\n")

distortions = plot_elbow_method(X_scaled)
silhouette = plot_silhouette_scores(X_scaled)
davies_bouldin = plot_davies_bouldin_scores(X_scaled)
dunn = plot_dunn_index(X_scaled)

# Resumen de resultados
print("\n" + "="*80)
print(" RESUMEN DE RESULTADOS")
print("="*80)

K_range = range(2, 10)
best_silhouette = K_range[np.argmax(silhouette)]
best_db = K_range[np.argmin(davies_bouldin)]
best_dunn = K_range[np.argmax(dunn)]

print(f"\n M茅todo del Codo: Sugiere k = 3 (punto de inflexi贸n)")
print(f" Coeficiente de Silueta: Mejor k = {best_silhouette} (score = {max(silhouette):.4f})")
print(f" ndice Davies-Bouldin: Mejor k = {best_db} (score = {min(davies_bouldin):.4f})")
print(f" ndice de Dunn: Mejor k = {best_dunn} (score = {max(dunn):.4f})")

print(f"\n CONCLUSIN: Los 4 m茅todos sugieren consistentemente k = 3 clusters,")
print(f"   lo cual es coherente con las 3 especies reales del dataset Iris.")
print("="*80 + "\n")

# Gr谩fico comparativo final
fig, ax = plt.subplots(figsize=(12, 7))

# Normalizar valores para comparaci贸n visual
from sklearn.preprocessing import minmax_scale

dist_norm = minmax_scale([distortions[i-1] for i in range(2, 10)])
sil_norm = minmax_scale(silhouette)
db_norm = minmax_scale([-x for x in davies_bouldin])  # Invertir para que mayor sea mejor
dunn_norm = minmax_scale(dunn)

ax.plot(K_range, dist_norm, marker='o', linewidth=2.5, markersize=10, 
        label='Codo (invertido)', color='#3498db')
ax.plot(K_range, sil_norm, marker='s', linewidth=2.5, markersize=10, 
        label='Silueta', color='#2ecc71')
ax.plot(K_range, db_norm, marker='^', linewidth=2.5, markersize=10, 
        label='Davies-Bouldin (inv)', color='#9b59b6')
ax.plot(K_range, dunn_norm, marker='D', linewidth=2.5, markersize=10, 
        label='Dunn', color='#e67e22')

ax.axvline(x=3, color='red', linestyle='--', linewidth=2.5, alpha=0.7, 
           label='k = 3 (贸ptimo)')

ax.set_xlabel('N煤mero de Clusters (k)', fontweight='bold', fontsize=13)
ax.set_ylabel('Valor Normalizado', fontweight='bold', fontsize=13)
ax.set_title('Comparaci贸n de los 4 M茅todos de Evaluaci贸n (Valores Normalizados)\nTodos los m茅todos sugieren k = 3', 
             fontweight='bold', fontsize=15, pad=20)
ax.grid(True, alpha=0.3, linestyle='--')
ax.legend(loc='best', framealpha=0.95, edgecolor='black', fontsize=11, ncol=2)
ax.set_xticks(K_range)

plt.tight_layout()
plt.show()

print(" An谩lisis completado exitosamente\n")
```

Los analisis de los indices nos muestran resultados similares en cuanto al numero de clusters optimos, los metodos apuntan que 2 seria un numero de clusters apropiado a pesar de que el dataset original tiene 3 clases. Tambien nos sugiere que aumentar el numero de clusters mas alla de 3 podria generar clusters de menor calidad con una separacion mas pronunciada entre ellos.

# K-Means con datos normalizados

**Parte 2: Clasificaci贸n de los tipos de flores basados en n煤mero k de clusters

y evaluaci贸n de resultados**

```python
# Importar las librer铆as necesarias para el an谩lisis
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, accuracy_score, adjusted_rand_score
from sklearn.decomposition import PCA
import seaborn as sns
from scipy.optimize import linear_sum_assignment

# Configuraci贸n de estilo mejorado
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = '#fafafa'
plt.rcParams['font.size'] = 11
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['figure.titlesize'] = 16

# Cargar dataset
iris = load_iris()

# Escalar caracter铆sticas
scaler = StandardScaler()
X_scaled = scaler.fit_transform(iris.data)

# Definir el n煤mero de clusters que decidimos tras la evaluaci贸n
num_clusters = 3  # Puedes cambiar este valor seg煤n los resultados del an谩lisis

# Aplicar k-means con el n煤mero seleccionado de clusters y m煤ltiples inicializaciones
kmeans = KMeans(n_clusters=num_clusters, n_init=50, random_state=42)
kmeans.fit(X_scaled)

# Predicciones de cluster del modelo
labels_pred = kmeans.labels_

# Visualizar las etiquetas originales y las predicciones
print("\n" + "="*70)
print(" RESULTADOS DEL CLUSTERING K-MEANS - DATASET IRIS")
print("="*70)
print(f"\nN煤mero de clusters: {num_clusters}")
print(f"\nEtiquetas reales:      {iris.target}")
print(f"Clusters predichos:    {labels_pred}")
print("="*70)

# Calcular la matriz de confusi贸n
conf_matrix = confusion_matrix(iris.target, labels_pred)

# Ajustar las etiquetas predichas para alinearlas con las etiquetas reales
def map_clusters_to_labels(conf_matrix):
    row_ind, col_ind = linear_sum_assignment(-conf_matrix)
    return col_ind

# Mapeo 贸ptimo de clusters a etiquetas
mapping = map_clusters_to_labels(conf_matrix)

# Ajustar las predicciones de clusters seg煤n el mapeo 贸ptimo
labels_pred_adjusted = np.array([mapping[label] for label in labels_pred])

# Calcular m茅tricas
overall_accuracy = accuracy_score(iris.target, labels_pred_adjusted)
ari_score = adjusted_rand_score(iris.target, labels_pred)

print(f"\n MTRICAS DE RENDIMIENTO:")
print(f"    Overall Accuracy: {overall_accuracy:.4f}")
print(f"    ARI (Adjusted Rand Index): {ari_score:.4f}")
print("="*70 + "\n")

# An谩lisis de precisi贸n del clustering
if num_clusters == 3:
    print(f" El algoritmo k-means con {num_clusters} clusters tiene una similitud")
    print(f"   con las etiquetas reales (ARI) de {ari_score:.4f}.")
else:
    print(f" El n煤mero de clusters seleccionado es {num_clusters}.")
    print(f"   Eval煤a si los resultados del clustering son coherentes con las especies reales.")

# =====================================
# VISUALIZACIONES MEJORADAS
# =====================================

# 1. MATRIZ DE CONFUSIN MEJORADA
fig, ax = plt.subplots(figsize=(9, 7))

# Crear heatmap con mejor dise帽o
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='RdYlGn', 
            linewidths=2, linecolor='white',
            cbar_kws={'label': 'Frecuencia', 'shrink': 0.8},
            annot_kws={'size': 14, 'weight': 'bold'},
            ax=ax, vmin=0, vmax=conf_matrix.max())

# Etiquetas con nombres de especies
especies = ['Setosa', 'Versicolor', 'Virginica']
ax.set_xticklabels(especies, rotation=0, ha='center', fontsize=11)
ax.set_yticklabels(especies, rotation=0, fontsize=11)

ax.set_xlabel('Clusters Predichos', fontsize=13, fontweight='bold', labelpad=10)
ax.set_ylabel('Especies Verdaderas', fontsize=13, fontweight='bold', labelpad=10)

# T铆tulo con m茅tricas
title_text = f'Matriz de Confusi贸n - K-means (k={num_clusters})\n'
title_text += f'Accuracy: {overall_accuracy:.3f} | ARI: {ari_score:.3f}'
ax.set_title(title_text, fontsize=15, fontweight='bold', pad=20)

# Agregar anotaciones adicionales
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        if i == j:
            ax.add_patch(plt.Rectangle((j, i), 1, 1, fill=False, 
                                      edgecolor='darkgreen', lw=3))

plt.tight_layout()
plt.show()

# 2. VISUALIZACIN PCA MEJORADA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Visualizaci贸n de Clustering K-means con PCA', 
             fontsize=16, fontweight='bold', y=1.02)

# Gr谩fico 1: Clusters predichos
scatter1 = axes[0].scatter(X_pca[:, 0], X_pca[:, 1], 
                          c=labels_pred, cmap='viridis', 
                          s=150, alpha=0.7, edgecolors='black', linewidth=1.5)
axes[0].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% varianza)', 
                   fontsize=12, fontweight='bold')
axes[0].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% varianza)', 
                   fontsize=12, fontweight='bold')
axes[0].set_title('Clusters Predichos por K-means', fontsize=13, fontweight='bold', pad=15)
axes[0].grid(True, alpha=0.3, linestyle='--')

# Agregar centroides
centroides_pca = pca.transform(kmeans.cluster_centers_)
axes[0].scatter(centroides_pca[:, 0], centroides_pca[:, 1], 
               marker='*', s=800, c='red', edgecolors='black', 
               linewidth=2, label='Centroides', zorder=5)

# Colorbar para el primer gr谩fico
cbar1 = plt.colorbar(scatter1, ax=axes[0])
cbar1.set_label('Cluster', fontsize=11, fontweight='bold')
axes[0].legend(loc='best', framealpha=0.9, edgecolor='black')

# Gr谩fico 2: Etiquetas reales
scatter2 = axes[1].scatter(X_pca[:, 0], X_pca[:, 1], 
                          c=iris.target, cmap='rainbow', 
                          s=150, alpha=0.7, edgecolors='black', linewidth=1.5)
axes[1].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% varianza)', 
                   fontsize=12, fontweight='bold')
axes[1].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% varianza)', 
                   fontsize=12, fontweight='bold')
axes[1].set_title('Especies Verdaderas (Ground Truth)', fontsize=13, fontweight='bold', pad=15)
axes[1].grid(True, alpha=0.3, linestyle='--')

# Colorbar para el segundo gr谩fico
cbar2 = plt.colorbar(scatter2, ax=axes[1], ticks=[0, 1, 2])
cbar2.set_label('Especie', fontsize=11, fontweight='bold')
cbar2.ax.set_yticklabels(['Setosa', 'Versicolor', 'Virginica'])

plt.tight_layout()
plt.show()

# 3. GRFICO ADICIONAL: Comparaci贸n lado a lado con leyendas
fig, ax = plt.subplots(figsize=(12, 8))

# Crear un gr谩fico combinado con s铆mbolos diferentes
colores_especies = ['#e74c3c', '#3498db', '#2ecc71']
especies_nombres = ['Setosa', 'Versicolor', 'Virginica']

for i in range(3):
    mask = iris.target == i
    ax.scatter(X_pca[mask, 0], X_pca[mask, 1], 
              c=colores_especies[i], s=200, alpha=0.6,
              edgecolors='black', linewidth=1.5,
              marker='o', label=f'Real: {especies_nombres[i]}')

# A帽adir contornos de clusters predichos
from matplotlib.patches import Ellipse
from scipy import stats

for cluster_id in range(num_clusters):
    mask_cluster = labels_pred == cluster_id
    points = X_pca[mask_cluster]
    
    if len(points) > 2:
        # Calcular elipse de confianza
        mean = points.mean(axis=0)
        cov = np.cov(points.T)
        
        # Eigenvalores y eigenvectores para la orientaci贸n
        eigenvalues, eigenvectors = np.linalg.eig(cov)
        angle = np.degrees(np.arctan2(eigenvectors[1, 0], eigenvectors[0, 0]))
        
        # Dibujar elipse
        ellipse = Ellipse(mean, width=np.sqrt(eigenvalues[0])*4, 
                         height=np.sqrt(eigenvalues[1])*4,
                         angle=angle, alpha=0.2, 
                         facecolor=f'C{cluster_id}',
                         edgecolor=f'C{cluster_id}', linewidth=2,
                         linestyle='--', label=f'Cluster {cluster_id}')
        ax.add_patch(ellipse)

# Centroides
centroides_pca = pca.transform(kmeans.cluster_centers_)
ax.scatter(centroides_pca[:, 0], centroides_pca[:, 1], 
          marker='X', s=500, c='black', edgecolors='yellow', 
          linewidth=3, label='Centroides K-means', zorder=10)

ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% varianza explicada)', 
              fontsize=13, fontweight='bold')
ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% varianza explicada)', 
              fontsize=13, fontweight='bold')

title_combined = f'Comparaci贸n: Especies Reales vs Clusters K-means (k={num_clusters})\n'
title_combined += f'Overall Accuracy: {overall_accuracy:.3f} | ARI: {ari_score:.3f}'
ax.set_title(title_combined, fontsize=15, fontweight='bold', pad=20)

ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.8)
ax.legend(loc='best', framealpha=0.95, edgecolor='black', fontsize=10, ncol=2)

plt.tight_layout()
plt.show()

print("\n Visualizaciones generadas exitosamente\n")
```

# Comparacion con otros metodos

```python
# Comparar "distancias" en clustering (Iris) midiendo coherencia y tiempo
# Nota: sklearn KMeans usa euclidiana; para Manhattan y Coseno usamos KMedoids (admite metric). 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from time import perf_counter
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids  # pip install scikit-learn-extra
from sklearn.metrics import confusion_matrix, accuracy_score, adjusted_rand_score
from scipy.optimize import linear_sum_assignment

# Configuraci贸n de estilo
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = '#f8f9fa'
plt.rcParams['font.size'] = 10
plt.rcParams['axes.labelsize'] = 11
plt.rcParams['axes.titlesize'] = 12
plt.rcParams['xtick.labelsize'] = 9
plt.rcParams['ytick.labelsize'] = 9

# -----------------------------
# Datos y escalado
# -----------------------------
iris = load_iris()
X = iris.data
y = iris.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

k = 3
random_state = 42

# -----------------------------
# Utilidades de evaluaci贸n
# -----------------------------
def hungarian_map_and_accuracy(y_true, labels_pred):
    cm = confusion_matrix(y_true, labels_pred)
    row_ind, col_ind = linear_sum_assignment(-cm)
    mapping = {pred_cluster: true_label for pred_cluster, true_label in zip(col_ind, row_ind)}
    labels_adj = np.array([mapping[c] for c in labels_pred])
    acc = accuracy_score(y_true, labels_adj)
    return cm, acc

def run_and_evaluate(model_name, model, X_use, y_true, fit_kwargs=None):
    if fit_kwargs is None:
        fit_kwargs = {}

    t0 = perf_counter()
    model.fit(X_use, **fit_kwargs)
    t1 = perf_counter()

    labels_pred = model.labels_
    cm, acc = hungarian_map_and_accuracy(y_true, labels_pred)
    ari = adjusted_rand_score(y_true, labels_pred)

    n_iter = getattr(model, "n_iter_", None)
    return {
        "modelo": model_name,
        "tiempo_s": t1 - t0,
        "iteraciones": n_iter,
        "overall_accuracy": acc,
        "ARI": ari,
        "confusion_matrix": cm,
    }

# -----------------------------
# Modelos a comparar
# -----------------------------
results = []

# 1) KMeans (Euclidiana)
kmeans = KMeans(n_clusters=k, n_init=50, random_state=random_state)
results.append(run_and_evaluate("KMeans\n(Euclidean)", kmeans, X_scaled, y))

# 2) KMedoids (Manhattan)
kmed_manh = KMedoids(n_clusters=k, metric="manhattan", init="k-medoids++", random_state=random_state)
results.append(run_and_evaluate("KMedoids\n(Manhattan)", kmed_manh, X_scaled, y))

# 3) KMedoids (Cosine)
kmed_cos = KMedoids(n_clusters=k, metric="cosine", init="k-medoids++", random_state=random_state)
results.append(run_and_evaluate("KMedoids\n(Cosine)", kmed_cos, X_scaled, y))

# 4) Spherical KMeans aproximado
X_unit = normalize(X_scaled)
kmeans_spherical = KMeans(n_clusters=k, n_init=50, random_state=random_state)
results.append(run_and_evaluate("KMeans\n(Aprox Cosine)", kmeans_spherical, X_unit, y))

# -----------------------------
# Tabla de resultados
# -----------------------------
df = pd.DataFrame([{k: v for k, v in r.items() if k != "confusion_matrix"} for r in results])
df = df.sort_values(by="ARI", ascending=False)

print("\n" + "="*80)
print(" COMPARACIN DE MTRICAS DE DISTANCIA EN CLUSTERING - DATASET IRIS")
print("="*80)
print(df.to_string(index=False))
print("="*80 + "\n")

# -----------------------------
# Visualizaciones mejoradas
# -----------------------------

# 1. Matrices de confusi贸n en una sola figura
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
fig.suptitle('Matrices de Confusi贸n por M茅trica de Distancia', 
             fontsize=16, fontweight='bold', y=0.995)

axes = axes.ravel()
colors = ['Blues', 'Greens', 'Purples', 'Oranges']

for idx, r in enumerate(results):
    sns.heatmap(r["confusion_matrix"], annot=True, cmap=colors[idx], fmt="d", 
                ax=axes[idx], cbar_kws={'label': 'Frecuencia'},
                linewidths=0.5, linecolor='gray')
    
    # T铆tulo con m茅tricas clave
    title = f"{r['modelo']}\nAccuracy: {r['overall_accuracy']:.3f} | ARI: {r['ARI']:.3f}"
    axes[idx].set_title(title, fontsize=11, fontweight='bold', pad=10)
    axes[idx].set_xlabel('Cluster Predicho', fontsize=10, fontweight='bold')
    axes[idx].set_ylabel('Etiqueta Real', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

# 2. Gr谩fico de barras comparativo - M茅tricas de rendimiento
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Comparaci贸n de Rendimiento entre M茅tricas de Distancia', 
             fontsize=16, fontweight='bold')

modelos_cortos = [r["modelo"].replace('\n', ' ') for r in results]
tiempos = [r["tiempo_s"] for r in results]
accuracies = [r["overall_accuracy"] for r in results]
aris = [r["ARI"] for r in results]

# Tiempo de ejecuci贸n
bars1 = axes[0].bar(range(len(results)), tiempos, color=['#3498db', '#2ecc71', '#9b59b6', '#e67e22'])
axes[0].set_xlabel('M茅todo', fontweight='bold')
axes[0].set_ylabel('Tiempo (segundos)', fontweight='bold')
axes[0].set_title('Tiempo de Ejecuci贸n', fontweight='bold')
axes[0].set_xticks(range(len(results)))
axes[0].set_xticklabels(modelos_cortos, rotation=15, ha='right')
axes[0].grid(axis='y', alpha=0.3)
for i, v in enumerate(tiempos):
    axes[0].text(i, v + max(tiempos)*0.02, f'{v:.4f}s', ha='center', fontsize=9)

# Overall Accuracy
bars2 = axes[1].bar(range(len(results)), accuracies, color=['#3498db', '#2ecc71', '#9b59b6', '#e67e22'])
axes[1].set_xlabel('M茅todo', fontweight='bold')
axes[1].set_ylabel('Accuracy', fontweight='bold')
axes[1].set_title('Overall Accuracy', fontweight='bold')
axes[1].set_xticks(range(len(results)))
axes[1].set_xticklabels(modelos_cortos, rotation=15, ha='right')
axes[1].set_ylim([0, 1.1])
axes[1].grid(axis='y', alpha=0.3)
for i, v in enumerate(accuracies):
    axes[1].text(i, v + 0.02, f'{v:.3f}', ha='center', fontsize=9)

# ARI (Adjusted Rand Index)
bars3 = axes[2].bar(range(len(results)), aris, color=['#3498db', '#2ecc71', '#9b59b6', '#e67e22'])
axes[2].set_xlabel('M茅todo', fontweight='bold')
axes[2].set_ylabel('ARI', fontweight='bold')
axes[2].set_title('Adjusted Rand Index', fontweight='bold')
axes[2].set_xticks(range(len(results)))
axes[2].set_xticklabels(modelos_cortos, rotation=15, ha='right')
axes[2].set_ylim([0, 1.1])
axes[2].grid(axis='y', alpha=0.3)
for i, v in enumerate(aris):
    axes[2].text(i, v + 0.02, f'{v:.3f}', ha='center', fontsize=9)

plt.tight_layout()
plt.show()

# 3. Gr谩fico de iteraciones
fig, ax = plt.subplots(figsize=(10, 6))
iteraciones = [r["iteraciones"] if r["iteraciones"] is not None else 0 for r in results]
bars = ax.bar(range(len(results)), iteraciones, color=['#3498db', '#2ecc71', '#9b59b6', '#e67e22'])

ax.set_xlabel('M茅todo', fontweight='bold', fontsize=12)
ax.set_ylabel('N煤mero de Iteraciones', fontweight='bold', fontsize=12)
ax.set_title('Iteraciones hasta Convergencia por M茅trica de Distancia', 
             fontweight='bold', fontsize=14, pad=15)
ax.set_xticks(range(len(results)))
ax.set_xticklabels(modelos_cortos, rotation=15, ha='right')
ax.grid(axis='y', alpha=0.3)

for i, v in enumerate(iteraciones):
    if v > 0:
        ax.text(i, v + max(iteraciones)*0.02, f'{v}', ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

# 4. Tabla visual resumida
fig, ax = plt.subplots(figsize=(12, 4))
ax.axis('tight')
ax.axis('off')

tabla_data = []
for r in results:
    tabla_data.append([
        r["modelo"].replace('\n', ' '),
        f"{r['tiempo_s']:.4f}",
        f"{r['iteraciones']}" if r['iteraciones'] else "N/A",
        f"{r['overall_accuracy']:.3f}",
        f"{r['ARI']:.3f}"
    ])

tabla = ax.table(cellText=tabla_data,
                colLabels=['M茅todo', 'Tiempo (s)', 'Iteraciones', 'Accuracy', 'ARI'],
                cellLoc='center',
                loc='center',
                colWidths=[0.3, 0.15, 0.15, 0.15, 0.15])

tabla.auto_set_font_size(False)
tabla.set_fontsize(10)
tabla.scale(1, 2.5)

# Colorear encabezados
for i in range(5):
    tabla[(0, i)].set_facecolor('#34495e')
    tabla[(0, i)].set_text_props(weight='bold', color='white')

# Colorear filas alternadas
for i in range(1, len(tabla_data) + 1):
    for j in range(5):
        if i % 2 == 0:
            tabla[(i, j)].set_facecolor('#ecf0f1')
        else:
            tabla[(i, j)].set_facecolor('#ffffff')

plt.title('Resumen Comparativo de M茅tricas de Distancia', 
          fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```
